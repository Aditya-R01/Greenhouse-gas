```python
numerical_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
```

```python
print("Checking for outliers in numerical columns using the IQR method:\n")

# Loop through each numerical column to check for outliers
for col in numerical_cols:
    Q1 = data_cleaned[col].quantile(0.25)
    Q3 = data_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data_cleaned[(data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)]

    if not outliers.empty:
        print(f"--- Outliers in '{col}' ---")
        print(f"Number of outliers: {len(outliers)}")
        print("Outlier values:\n")
        print(outliers[[col]].to_markdown(index=False, numalign="left", stralign="left"))
        print("\n")
    else:
        print(f"No outliers found in '{col}'\n")
```

```python
original_shape = data_cleaned.shape
print(f"Original DataFrame shape: {original_shape}")
```

```python
# Columns identified with outliers
outlier_cols = [
    'Supply Chain Emission Factors without Margins',
    'Margins of Supply Chain Emission Factors',
    'Supply Chain Emission Factors with Margins'
]

# Initialize a boolean mask for identifying outliers
# Start with a mask that assumes no rows are outliers
# We will then update this mask to true for rows that contain outliers
outlier_mask = pd.Series([False] * len(data_cleaned), index=data_cleaned.index)

# Loop through each identified column to find outliers and update the mask
for col in outlier_cols:
    Q1 = data_cleaned[col].quantile(0.25)
    Q3 = data_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Update the outlier_mask to True for rows that are outliers in the current column
    # Use OR (|) to combine masks, so if a row is an outlier in any of the columns, it's marked
    outlier_mask = outlier_mask | (data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)

# Filter the DataFrame to remove rows identified as outliers
data_cleaned1 = data_cleaned[~outlier_mask].copy() # Using .copy() to avoid SettingWithCopyWarning

cleaned_shape = data_cleaned1.shape
print(f"DataFrame shape after outlier removal: {cleaned_shape}")
print(f"Number of rows removed: {original_shape[0] - cleaned_shape[0]}\n")

print("First 5 rows of the DataFrame after outlier removal:")
data_cleaned1.head(5)
```
